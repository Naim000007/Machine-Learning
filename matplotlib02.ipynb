```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
print(y)
print(z)
```

```python
# ploting data
#sine wave
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
```

```python
# adding title x-axis and y axis and labels
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel(" sine value")
plt.title("sin wave")
```

```python
#parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```
